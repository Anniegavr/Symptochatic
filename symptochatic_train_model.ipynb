```python
X_new = np.array(X_list)
```

```python
new_X = []
```

```python
#creates an array that excludes white spaces
for i in range(len(X_new)):
    new_X.append([])
    new_X[-1].append(X_new[i, 0])
    new_X[-1].append(
        ' '.join(X_new[i, 1:])
    )
    new_X[-1][1] = new_X[-1][1].strip()
```
